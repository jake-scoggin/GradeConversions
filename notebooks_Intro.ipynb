# UConn International GPA Conversion Project

This project lets you convert international GPAs to a USA scale. No process is perfect - take the results with a grain of salt.

```python
# Enter grade as a number between 0 and 100:
grade = 45

import numpy as np
import matplotlib.pyplot as plt

def get_fit(x1, y1, x2, y2):
    """Generates and plots a lienar fit between (x1, y1) and (x2, y2).
    Returns the linear polynomial used to make the fit.
    """
    x = np.array([x1, x2])
    y = np.array([y1,  y2])
    coeff = np.polyfit(x, y, 1)
    poly = np.poly1d(coeff)
    plt.plot(x, poly(x), linestyle='--')
    return poly

polyA = get_fit(60, 3.7, 100, 4.0) # A range
polyB = get_fit(45, 2.7, 60,  3.7) # B range
polyC = get_fit(33, 1.7, 45,  2.7) # C range
polyF = get_fit(0,    0, 33,  1.7) # D range

#
if (grade > 0) and (grade < 33):
    poly = polyF
elif (grade >= 33) and (grade < 45):
    poly = polyC
elif (grade >= 45) and (grade < 60):
    poly = polyB
elif (grade >= 60) and (grade <= 100):
    poly = polyA
else:
    raise RuntimeError(f"Grade {grade} is not between 0 and 100.")

equivalent_gpa = poly(grade)
print(f"The equivalent grade is {equivalent_gpa:.3f}.")

plt.scatter(grade, equivalent_gpa)
plt.show()   
```
